```python
data_exp = data

ratehtml = BeautifulSoup(data_exp.text, 'html.parser')

#td class="details" contains each entry segment in the page
rateentry = ratehtml.select('tr.expand_collapse')

#dictionary and lists to store scrape
wargame_test = {}
position_list = []
title_list = []
platform_list = []
pub_date_list = []
rating_list = []

#identify html location relative to start of scrape segment tr.expand_collapse and iterate through page
for entry in rateentry:
    
    #select relative html location, retrieve info as text and add it to appropriate list
    #strip white space and /n where present and remove full stops and convert to integers
    position = entry.select('td.details a span.title.numbered')[0].get_text()
    position_list.append(int(position.strip()[0:-1]))
                        
    title = entry.select('td.details h3')[0].get_text()
    title_list.append(title)
    
    platform = entry.select('td.details div.platform span.data')[0].get_text()
    platform_list.append(platform.strip())
    
    pub_date = entry.select('td.details span')[3].get_text()
    pub_date_list.append(pub_date)
    
    rating = entry.select('td.score a.metascore_anchor div')[0].get_text()
    rating_list.append(int(rating))

#Store output in dictionary and view    
wargame_test = {"position": position_list, "title": title_list, "platform": platform_list, "pub_date": pub_date_list, "rating": rating_list}
print(wargame_test)
```

```python
#convert to dataframe to clean up
mc_rating = pd.DataFrame.from_dict(wargame_test)
#added .strip and removed full stop in position

#label with game genre
mc_rating['genre'] = 'wargame'

#convert date format for use
mc_rating['pub_date'] = pd.to_datetime(mc_rating['pub_date'])

mc_rating
#add pages and genres
#export
```

```python
#export to csv
mc_rating.to_csv('wargame_MetaCritScores.csv')
```
