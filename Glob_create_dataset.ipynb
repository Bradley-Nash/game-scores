```python
#IMPORTING PACKAGES

#glob to compile scraped dataset from csv file exports
import glob

#pandas to visualise glob success
import pandas as pd
```

```python
#GLOB FILES

#get list of webscrape batch result csv files in the working directory as a list
files = glob.glob('*.{}'.format('csv'))
files
```

```python
#COMPILE DATAFRAMES

#create composite dataset by stitching each csv together

#assign empty dataframe to iteratively add to 
mcs = pd.DataFrame()

#loop through csvs in glob list
for file in files:
    
    #import csv and overide temporary variable dataframe
    temp_df = pd.read_csv(file)
    
    #Add the temporary variable dataframe to the bottom of the storage dataframe
    mcs = pd.concat([mcs, temp_df], axis=0, ignore_index = True)
    
#drop column 1 which is the unnamed index from csv exporting
mcs = mcs.iloc[:,1:]
mcs
```

```python
#CHECKING COMPLETEDNESS

#Use info method to look for equal non-null data and correct data types
mcs.info()

#Note: pub_date column is stored as an object
```

```python
#PREPARATION

#Convert date-time column to data-time datatype
mcs['pub_date'] = pd.to_datetime(mcs['pub_date'])

#Check pub_date column is stored as datetime
mcs.info()

#Note: datetime assignment will be lost on csv export
```

```python
#Export globbed csv to an output folder and remove pandas artificial index
mcs.to_csv('./output/MetaCritScores.csv', index=False)
```
